```python
# Import Required Libraries

import pandas as pd
import re
import matplotlib.pyplot as plt
```

```python
# import the log file and make a dataframe to store it

filename = "geldropseweg.txt"
Wifi = pd.DataFrame()
```

```python
count=0   # count for total entries
with open (filename, "r") as file:               # open log file

    for line in file:                            # read line by line
        '''
        Search through the [light,read] log to get the details of the nodes.
        This approach assumes that the data will be in order, otherwise Friendly_name could be different
        '''
        if re.search("Cell", line):                                 # search all parameters
            
            Address = re.search(r'.*Address: (.*)', line)
            
            Wifi.loc[count,'Address'] = Address.group(1)
            
        if re.search("ESSID", line):
            
            Essid = re.search(r'.*: (.*)', line)
            
            Wifi.loc[count,'ESSID'] = Essid.group(1)
        if re.search("Mode",line):
            
            Channel = re.search(r'.*Channel: (.*)',line)
            Wifi.loc[count,'Channel']= Channel.group(1)
            
        if re.search("Signal",line):
            
            Signal = re.search(r'.*Signal: (.*) dB.*',line)
            Wifi.loc[count,'Signal']= Signal.group(1)
        
        if re.search("Encryption",line):
            
            Encrypt = re.search(r'.*Encryption: (.*)',line)
            Wifi.loc[count,'Encryption'] = Encrypt.group(1)
 
            count = count + 1

#Wifi = Wifi.sort_values(by=['Signal'],ascending=True)
print(Wifi)
```
